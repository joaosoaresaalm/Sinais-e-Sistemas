**QUESTÃO 14**

```python
ant = 0
new_list = []
t = np.linspace(-10, 10, 500)
y = ss.square(t, duty=.5)

for i in y:
    if i != ant and ant != 0:
        for j in range(20):
            new_list.append(0)
    new_list.append(i)
    ant = i

t = np.linspace(-10, 10, len(new_list))
plt.plot(t, new_list)
```

**QUESTÃO 17**

```python
def rampa(t): 
    return np.maximum(0, t)

def sat(t, valor1, valor2):
    if t <= valor1:
        return valor1
    elif t >= valor2:
        return valor2
    else:
        return t
    
limiar_inferior, limiar_superior = 2, 5
t = np.linspace(0, 10, 500)
y = rampa(t)
y = [sat(i, limiar_inferior, limiar_superior) for i in y]
plt.plot(t, y)
```

Foi utilizado uma implementação de uma função de saturação
